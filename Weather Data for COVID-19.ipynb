# 1.How do confirmed cases and fatalities vary over time for different regions?

```python
data=pd.read_csv(r"training_data_with_weather_info_week_1.csv")
```

```python
g = sns.FacetGrid(data, col='Country/Region',col_wrap=5,sharex=False, sharey=False)
g.map(sns.lineplot, 'ConfirmedCases', 'Fatalities')
```

# 2.Which countries or regions have the highest and lowest total confirmed cases and fatalities?

```python
totals = data.groupby('Country/Region')[['ConfirmedCases', 'Fatalities']].sum()
totals.sort_values(by='ConfirmedCases', ascending=False)
totals.sort_values(by='Fatalities', ascending=False)
```

- highest comfirmed cases and fatalities are in china
- lowest confirmed cases and fatalities are in Zambia
